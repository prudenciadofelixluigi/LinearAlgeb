a matrix where it shows an empty set. In this cell when the cell is defined and the program understand that the given commands does not contains any elements it will display the matrix is null otherwise it is false. 

image.png

```python
def mat_desc(matrix):
    if matrix.size > 0:
        is_empty = True if matrix.shape[0] == matrix.shape[1] else False 
        print(f'Matrix:\n{matrix}\n\nShape:\t{matrix.shape}\nRank:\t{matrix.ndim}\nIs empty: {is_empty}\n')

    else:
        print('Matrix is Null')
```

```python
edi_empty = np.array([
    [],
    [],
    []
])

edi_hindi_empty = np.array([
    [9, 7, 8],
    [1, 3, 4]
])
mat_desc(edi_empty)
mat_desc(edi_hindi_empty)
```
